```python
import sympy
```

```python
e1 = sympy.symbols('epsilon_1')
display(e1)
```

```python
e2 = sympy.symbols('epsilon_2')
display(e2)
```

```python
ec1 = sympy.Eq(0.5,2-e1-e2)
display(ec1)
```

```python
ec2=sympy.Eq(4.9,3*e1+4*e2)
display(ec2)
```

```python
sympy.solve([ec1,ec2],[e1,e2])
```

```python
nt=5+2*x1+2*x2
```

```python
ymetano=(2-x1-x2)/nt
yagua=(3-x1-2*x2)/nt
yCO=x1/(nt)
yco2=x2/(nt)
yh2=(3*x1+4*x2)/nt
```

```python
print(f"ymetano = {ymetano}")
print(f"yagua = {yagua}")
print(f"yCO = {yCO}")
print(f"yco2 = {yco2}")
print(f"yh2 = {yh2}")
```
