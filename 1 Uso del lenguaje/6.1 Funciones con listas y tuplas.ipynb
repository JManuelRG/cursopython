```python
# Ejemplos:
#Definir lista y tupla

lista = [0,4,3,2,7,9,86,12]
tupla=(10,20,30,40,50,50,70,60, 40, -12, 0)

#longitud de ambas:
print(len(lista))
print(len(tupla))
```

```python
#Valores minimos y máximos

print("Valor Mínimo de lista: ", min(lista))
print("Valor máximo de lista: ", max(lista))

print("Valor Mínimo de tupla: ", min(tupla))
print("Valor máximo de tupla: ", max(tupla))
```

```python
# Ejemplo: Obtener los valores de la función Y=X/2
# Agregar elemento con append para construir una lista con el comando for, se genera un vector de datos

valoresX = [] # Lista vacía

for i in range(1,41): #Agregar 40 elementos a la lista uno a uno
    valoresX.append(i)
    
print(valoresX)
```

```python
# Operaciones con la lista (vector), para obtener Y dividir cada elemento de X entre 2 uno a uno

valoresY=[0]*len(valoresX) #vector vacío
for i in range(len(valoresX)):
    valoresY[i]=valoresX[i]/2
    
print(valoresY)
```
