```python
# Ejemplos:
#Definir lista y tupla

lista = [0,4,3,2,7,9,86,12]
tupla=(10,20,30,40,50,50,70,60, 40, -12, 0)

#longitud de ambas:
print(len(lista))
print(len(tupla))
```

```python
#Valores minimos y mximos

print("Valor Mnimo de lista: ", min(lista))
print("Valor mximo de lista: ", max(lista))

print("Valor Mnimo de tupla: ", min(tupla))
print("Valor mximo de tupla: ", max(tupla))
```

```python
# Ejemplo: Obtener los valores de la funcin Y=X/2
# Agregar elemento con append para construir una lista con el comando for, se genera un vector de datos

valoresX = [] # Lista vaca

for i in range(1,41): #Agregar 40 elementos a la lista uno a uno
    valoresX.append(i)
    
print(valoresX)
```

```python
# Operaciones con la lista (vector), para obtener Y dividir cada elemento de X entre 2 uno a uno

valoresY=[0]*len(valoresX) #vector vaco
for i in range(len(valoresX)):
    valoresY[i]=valoresX[i]/2
    
print(valoresY)
```
