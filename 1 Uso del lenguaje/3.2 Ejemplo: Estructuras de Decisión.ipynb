Estructura de decisin if:



Donde "c" es una comparacin lgica, que arroja un valor Verdadero o Falso. Si c es verdadero se ejecutar el bloque de instrucciones.

```python
a=3
b=7

print(a==b) #Igual que
print(a<b) #Menor que
print(a>b) #Mayor que
print(a>=b) #Mayor o igual que
print(a<=b) #Menor o igual que
print(a!=b) #Diferente que
```

```python
# Entrada de datos
# Estructura de decisin
"""
Comando input para entrada de datos

Ejercicio: Crear una calculadora que lea dos nmeros y obtenga la suma de ambos
"""

print("Este programa obtiene la operacin seleccionada entre dos nmeros.")
#Lectura de nmeros:
num1 = float(input("Introduce el primer nmero = "))

num2 = float(input("Introduce el segundo nmero ="))
op = str(input("Introduce (S) para suma, (R) para resta, (D) para divisin, (M) para multiplicacin, (E) para potencia = \n"))

if op == "S":
    print(f"La suma de {num1} + {num2} = {num1+num2}")
elif op =="M":
    print(f"La multiplicacin de {num1} * {num2} = {num1*num2}")
elif op =="R":
    print(f"La resta de {num1} - {num2} = {num1-num2}")
elif op =="D":
    print(f"La divisin de {num1} / {num2} = {num1/num2}")
elif op =="E":
    print(f"La exponenciacin de {num1} elevado a la {num2} = {num1**num2}")
```
