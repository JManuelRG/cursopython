FUNCIONES LAMBDA.

Son funciones anónimas definidas por el usuario.

Una función anónima, como su nombre indica es una función sin nombre. 
Python podemos ejecutar una función sin definirla con def. 
De hecho son similares pero con una diferencia fundamental:

El contenido de una función lambda debe ser una única expresión en lugar de un bloque de acciones.

```python
# Ejemplo

# Construir una función que triplique un número, con def la sintaxis sería:

def triplicar(num):
    return 3*num

print(triplicar(4))
```

```python
# Con una función lambda:

# Sintaxis: variable = lambda entrada:operaciones

triple_con_lambda = lambda num:num*3 # no requiere un nombre, sí una variable

print(triple_con_lambda(3))
```

```python
# Principales usos de las funciones lambda:

# Sustituir valores en una función

# Ejemplo: Sea la función f(x) = x^2, encontrar los valores de la función para x = 0 al 10.

# Primero definir la función:
f=lambda x:x**2

# Sustituir x = 0 en la función:
print(f(0))

# Sustituir x = 1 en la función:
print(f(1))
```

```python
# Así sucesivamente, es más fácil realizar un ciclo for para generar los valores del 0 al 10
print("Tabla de Valores.")
print("%15s %15s" %("|Valores de X|", "|Valores de F(X)|"))

# calcular valores de f(x) para cada x
for i in range(0,11):
    print("%15d %15d" %(i, f(i)))
```

NOTA: Se utilizó un formato personalizado para mostrar la tabla con el comando print:


```python
# FUNCIONES LAMBDA DE VARIAS VARIABLES O ARGUMENTOS

# Ejemplo: Sea la función: g(a,b) = a + b
# Definir la expresión como una función lambda:

g = lambda a, b: a + b

print(g(1,2))
print(g(10,40))
print(g(100,200.2))
```
