# Estrutura for

## Sintaxis

para (un identificador, o una variable) en (valores=rango):

  bloque de instrucciones

## Para el rango
Se usa la función o método range(valor), que devuelve una lista de valores, dependiendo del valor que se use como argumento.

### range(10)
Devuelve los números del 0 al 9

### range(valor inicial, valor final)
Devuelve los números desde el valor inicial hasta el valor final menos uno.


```python
# Ejemplo
# Contar de 1 al 10

# De manera secuencial:
print(1)
print(2)
print(3)
print(4)
print(5)
print(6)
print(7)
print(8)
print(9)
print(10)
```

```python
# Para hacerlo con la estructura:
# range(inicio, final)

for i in range(1,11): # no incluye el valor final
  print(i)
```

```python
# Para hacerlo con la estructura, contando de 2 en 2
# range(inicio, final, paso)

for i in range(2,11,2): # no incluye el valor final
  print(i)
```
