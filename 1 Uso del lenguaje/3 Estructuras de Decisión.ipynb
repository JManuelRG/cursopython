Estructura de decisión if:



Donde "c" es una comparación lógica, que arroja un valor Verdadero o Falso. Si c es verdadero se ejecutará el bloque de instrucciones.

```python
a=3
b=7

print(a==b) #Igual que
print(a<b) #Menor que
print(a>b) #Mayor que
print(a>=b) #Mayor o igual que
print(a<=b) #Menor o igual que
print(a!=b) #Diferente que
```

```python
# Entrada de datos
# Estructura de decisión
"""
Comando input para entrada de datos

Ejercicio: Crear una calculadora que lea dos números y obtenga la suma de ambos
"""

print("Este programa obtiene la operación seleccionada entre dos números.")
#Lectura de números:
num1 = float(input("Introduce el primer número = "))

num2 = float(input("Introduce el segundo número ="))
op = str(input("Introduce (S) para suma, (R) para resta, (D) para división, (M) para multiplicación, (E) para potencia = \n"))

if op == "S":
    print(f"La suma de {num1} + {num2} = {num1+num2}")
elif op =="M":
    print(f"La multiplicación de {num1} * {num2} = {num1*num2}")
elif op =="R":
    print(f"La resta de {num1} - {num2} = {num1-num2}")
elif op =="D":
    print(f"La división de {num1} / {num2} = {num1/num2}")
elif op =="E":
    print(f"La exponenciación de {num1} elevado a la {num2} = {num1**num2}")
```
