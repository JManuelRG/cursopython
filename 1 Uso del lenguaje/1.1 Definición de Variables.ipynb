# Características de variables en Python
- Pueden contener cualquier nombre

- No pueden definirse con un nombre que empiece con un número, pero sí con combinaciones de texto, números o símbolos.


```python
# Definir una variable entera
var_entera = 10
print(var_entera)
type(var_entera)
```

```python
# Definir una variable flotante
var_flotante = 45.55
print(var_flotante)
type(var_flotante) 
```

```python
# Los nombres de variables no pueden comenzar con un número
1var = 12
```

```python
# Pueden finalizar con un número
var_flotante2 = 3.14159265358979
print(var_flotante2)
type(var_flotante2)
```

# Definir Números Complejos

Los números complejos se definen de la forma:

$ Número = a + b(i) $

Dónde:

$ i = \sqrt(-1)  $

Con el comando **Complex**.
Por ejemplo, definamos:
$num = 1 + 5*i $

Para python, la parte imaginaria $i$, le llama $j$.

```python
# Número complejo:
num_complejo = complex(1,5)
print(num_complejo)
type(num_complejo)
```

```python
# Llamar el valor real
print(num_complejo.real)
print(num_complejo.imag)
```

# Definir Variables (Boolean) Lógicas

Son variables binarias (que solo aceptan dos valores):

**Verdadero (True)**

**Falso (False)**

```python
# Asignar valor entre True o False
var_bin1 = False
var_bin2 = True 

print(var_bin1)
print(var_bin2)
type(var_bin1)
type(var_bin2)
```
