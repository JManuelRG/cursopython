Sirve para repetir instrucciones de manera controlada, un número de veces conocido


```python
""""
Estructuras de Control

For: Para

"""

# FOR
# Imprimir los números del 1 al 10

"""
 SINTAXIS
for identificador in range(inicial, final, *incrementos):
    instrucciones
    
    
el identificador es entero

"""

# Ejemplo: Imprimir los números del 1 al 10
for i in range(1,11):
    print(i)
```

```python
# Ejemplo Imprimir los números del 0 al 20
for i in range(21):
    print(i)
    
#Python numera desde 0 hasta i
```

```python
# Ejemplo: Imprimir los números del 10 al 25
for i in range(2,26):
    print(i)
```

```python
# Ejemplo: Imprimir los números del 10 al 40 de 2 en 2
for i in range(10,41,2):
    print(i)
```

```python
# Imprimir el cuadrado de los números del 1 a 20

for i in range(1,21):
    print(f"El cuadrado de {i} es: {i**2}")
```

```python
#Construir Tabla Ordenada
# con los cuadrados del 1 al 20
#Encabezado:
print("%3s %6s %10s" %("i", "Número", "Cuadrado"))
for i in range(1,21):
    print("%3d %6d %10.5f" %(i, i, i**2))
    

#d=entero
#f=flotante
#s=cadena de caracteres
```
