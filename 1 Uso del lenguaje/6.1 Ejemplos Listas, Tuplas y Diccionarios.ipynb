```python
# LISTAS: SON VECTORES O ARREGLOS LINEALES DE DATOS
# PUEDEN CONTENER DIFERENTES TIPOS DE DATOS
# SON INDEXABLES
# SON MUTABLES
# EJEMPLOS (se declaran con corchetes [])

lista1 = [20, 3.1416, "Texto", True] # lista heterogénea
print(lista1)

lista2 = [0,2,4,6,8,10] # Lista homogénea
print(lista2)
```

```python
# INDEXACIÓN DE LISTAS
# Las listas se indexan partiendo desde 0, es posible llamar por separado cada elemento y realizar operaciones 
print(lista1[0])
print(lista1[1])
print(lista1[2])
print(lista1[3])
```

```python
# Las listas son mutables, podemos cambiar los valores de cualquier elemento
print(lista1)
lista1[0]= "Valor Cambiado"
print(lista1)
```

```python
""""
TUPLAS:
# PUEDEN CONTENER DIFERENTES TIPOS DE DATOS
# SON INDEXABLES
# NO SON MUTABLES
# EJEMPLOS (se declaran con paréntesis)

"""

tupla1 = (20, 25, 30, 35, 40)

print(tupla1)
```

```python
# Las tuplas son indexables también
print(tupla1[0])
print(tupla1[1])
print(tupla1[2])
print(tupla1[3])
print(tupla1[4])
```

```python
# No es posible cambiar sus valores

tupla1[0]=1 # Genera un error
```

```python
# Los elementos de tuplas o listas son iterables

for i in lista1:
    print(i)
```

```python
# Es posible hacer operaciones con cada elemento de la lsta o tupla
for i in tupla1:
    print(i+0.5)
```

```python
"""
DICCIONARIOS.
SON PARECIDOS A LAS LISTAS, CON LA DIFERENCIA DE QUE ACEPTAN ÍNDICES PERSONALIZADOS LLAMADOS CLAVES.
SE DECLARAN ENTRE LLAVES: {}, las claves deben ser cadenas de caracteres, los valores pueden ser cualquier tipo de dato.

"""

diccionario1 = {"clave1":24, "clave2":30, "clave3":"45"}
print(diccionario1)
```

```python
# Son indexables, los índices son las claves:
print(diccionario1["clave1"])
print(diccionario1["clave2"])
print(diccionario1["clave3"])
```

```python
# Son útiles para agrupar objetos, ejemplo
propiedades = {"Sustancia":"Agua", "Tebullicion":100, "Tfusion":0, "densidad":999, "Cp":1}
print(propiedades)
```

```python
# Es posible iterar entre los elementos del diccionario

for key in propiedades:
    print(f'La clave {key} del diccionario "propiedades" es:  {propiedades[key]}')
```
