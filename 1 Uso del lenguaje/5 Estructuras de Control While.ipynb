Estructuras de Control: While y For


```python
"""
ESTRUCTURA FOR
Itera un nmero definido de pasos.

"""
#Ejemplo: Imprimir los nmeros de 1 al 10

for i in range(1,11): # Sin incluir el 11
    print(i)
```

```python
#Ejemplo: Imprimir los nmeros de 1 al 10

for i in range(1,21): # Sin incluir el 11
    print(i)
```

```python
""" 
While: Repita una serie de pasos mientras una condicin se cumpla, es decir, sea verdad (True)
No se conoce previamente el nmero de iteraciones, pero se detiene cuando se deja de cumplir la condicin.

"""
# Cuenta hasta 10

n=1
while n<11:
    print(n)
    n+=1 # Equivale a n=n+1, incrementa el valor de n
```

```python
# Ejemplo: Sucesin de Fibonacci con 20 trminos
N=20
a,b=0,1 #Doble asignacin de variables
i = 1 #Iteraciones

print("%4s %4s" %("Trmino","Valor")) # Encabezado de tabla
while i<N:
    print("%4d %4d" %(i,b)) # Elemento i de la tabla
    a,b = b, a+b
    i+=1
```

```python
# Es posible salir de las iteraciones de los ciclos for o while con el comando break: o saltarse esa iteracin con el comando continue:

print("Para FOR:")
for i in range(1,100):
    print(i)
    if i==5:
        break
```

```python
print("Para WHILE:")
n=1
while n<200:
    print(n)
    n+=1
    if n==10:
        break
```

```python
# Saltarse una iteracin con continue:


print("Para FOR:")
for i in range(1,11):
    if i==6:
        continue
    print(i)


```

```python
print("Para WHILE:")
n=0
while n<10:
    n+=1
    if n==3:
        continue
    print(n)
    
```
