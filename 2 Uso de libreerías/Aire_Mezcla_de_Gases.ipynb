# Soluci√≥n:

#$$\dot{Q}/{n} =R \int_{T_1}^{T_2} \frac{C_P}{R} dT$$

# a) Usando la tabla del polinomio de Cp

```python
import sympy #Symbolic Python

T = sympy.symbols('T')

CpO2 = 3.639+.506e-3*T-.227e5*T**(-2)
CpN2 =3.28+.593e-3*T+0.04e5*T**(-2)
```

```python
display(CpO2)
```

```python
display(CpN2)
```

# Capacidad Calor√≠fica de la Mezcla (Aire):
$$(C_P/R)_M = \sum_i{(C_P/R)_iy_i}$$

$$\frac{Cp}{R}_{Mezcla}=\frac{Cp}{R}_{O} y_{O} +\frac{Cp}{R}_{N} y_N$$

```python
CpM = 0.21*CpO2 + 0.79*CpN2
print("Cp/R para el aire:")
display(CpM)
```

#$$\frac{\dot{Q}}{\dot(n)}=R\int_{320 K}^{260 K}{\frac{C_P}{R}_{Aire}dT}$$

```python
solucion = 8.314*sympy.integrate(CpM, (T, 320, 260))
print(solucion) # En J/mol
```

a) Resultado = 
$$\dot{Q}/\dot{n} = -1747.31 J/mol \rightarrow kJ/kmol$$
