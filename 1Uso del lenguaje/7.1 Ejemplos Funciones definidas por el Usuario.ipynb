# Funciones definidas por el usuario

Es una forma de agrupar una serie de instrucciones.

image.png

```python
# Ejemplo: Función que devuelve la suma de dos números
def suma(numero1,numero2):
  global numero3
  numero3 = numero1+numero2
  return numero3
```

```python
# Llamando la función con argumentos por posición
# El primer argumento asume que es numero1, el segundo numero2
print(suma(100,200))
print(suma(400,500))
print(suma(3.4,4.444))

# O concatenar
print(suma("Hola ", "Mundo"))
# print(suma("F",4)) no se puede concatenar
```

# Funciones Recursivas

Una función que se llama a sí misma dentro del bloque de código.


```python
# Ejemplo: Cuenta Regresiva:
def cuenta_regresiva(numero):
  if numero>0:
    print(numero)
    numero-=1 # numero = numero-1
    cuenta_regresiva(numero)
  else:
      print(numero, "Finalizado")

cuenta_regresiva(10)
```

# Factorial de un número

# $$n! = n (n-1)(n-2)...(1) $$

```python
# Funciones recursivas, se ejecutan dentro de sí mismas
# Ejemplo: Obtener el factorial de un número positivo 

def factorial(numero):
    if numero==0:
      numero=1
    elif numero>1: 
        numero=numero*factorial(numero-1)
    return numero
            
print(factorial(4))
print(factorial(1))
print(factorial(13))
print(factorial(0))
```
