```python
from sympy import *
# Definir T como variable algebraica o simbólica
T=symbols('T')
display(T)
```

```python
# Definir polinomios
CpAcL = 0.123+18.6e-5*T # Acetona líquida
CpAcv= 0.07196+20.1e-5*T-12.78e-8*T**2+34.76e-12*T**3# Acetona vapor

print("Polinomio Cp para acetona líquida en kJ/mol°C, T en °C:")
display(CpAcL)
print("")
print("Polinomio Cp para acetona vapor kJ/mol°C, T en °C:")
display(CpAcv)
```

```python
# Datos conocidos
Vac= 0.0734 # Volumen específico molar de Ac líquida en L/mol
Tbac=56 #°C Teb de acetona a 1 atm
dHevac= 30.2 # Entalpía de evaporación de acetona L->v en kJ/mol
```

```python
H1 = Vac*(1-5)+integrate(CpAcL, (T,20,Tbac))+dHevac+integrate(CpAcv, (T, 56, 65))
print(f"H1 = {Vac*(1-5)} + {integrate(CpAcL, (T,20,Tbac))} + {dHevac} + {integrate(CpAcv, (T, 56, 65))} J/mol")
print(f"El valor de H1 es {H1} kJ/mol")
```

```python
H3=Vac*(1-5)+integrate(CpAcL,(T,20,56))+dHevac+integrate(CpAcv,(T,56,20))
print(f"H3 = {Vac*(1-5)} + {integrate(CpAcL,(T,20,56))} + {dHevac} + 0 + {integrate(CpAcv,(T,56,20))}  kJ/mol")
print(f"El valor de H3 es {H3} kJ/mol")
```
