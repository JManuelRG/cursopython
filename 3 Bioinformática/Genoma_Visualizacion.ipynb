```python
import matplotlib.pyplot as plt
import numpy as np

# Secuencia
sequence = "CTTGGTTCATTTAGAGGAAGTAAAAGTCGTAACAAGGTTTCCGTAGGTGAACCTGCGGAAGGATCATTAAAATATACTTATTACACTGTTTTGGAAGCAAACCACATACCAATTTAAATTTCAATTTCTTTAAAAAATCAAAAACTTTCAACAACGGATTCTTGGTTCTCGCATCGATGAAGGAACGCAGCGAATTGCGATACGTATCATGACTTGCAGACGTGAATCATCAATCTTTGAACGCACATTGCGCCTTGAGGTATTCCTCAAGGCATGCCTGTTTGAGCGTCGGCTCCCTTCAAACCCCCGGGTTTGGTGTTGCCTTCCGAAATATCACAGTTGTCGCAATACGTTACTTCAACTTTATTCTTTCGCCCTCAAATCAGGTAGGACTACCCGCTGAACTTAAGCATATCAATAAGCGGAGGAA" * 10  # 1290 caracteres

# Configuración
nucleotides = ['A', 'C', 'T', 'G']
nucleotide_colors = {'A': 'lightskyblue',
                     'T': 'khaki',
                     'C': 'forestgreen',
                     'G': 'lightpink'}
y_positions = {n: i + 1 for i, n in enumerate(nucleotides)}  # Asignar valores a A, C, T, G
line_length = 100  # Cantidad de nucleótidos por línea

# Dividir la secuencia en líneas
lines = [sequence[i:i + line_length] for i in range(0, len(sequence), line_length)]
num_lines = len(lines)

# Crear el gráfico
plt.figure(figsize=(20, len(lines) * 1.5))  # Ajustar tamaño según número de líneas
for line_num, line in enumerate(lines):
    for pos, nucleotide in enumerate(line):
        x = pos
        y = len(lines) - line_num  # La línea más alta tiene el mayor valor de Y
        color = nucleotide_colors.get(nucleotide, 'gray')  # Por si hay caracteres no esperados
        # Dibujar cuadro
        plt.gca().add_patch(plt.Rectangle((x - 0.5, y - 0.5), 1, 1, color=color, edgecolor='black'))
        # Etiquetar cuadro con el nucleótido
        plt.text(x, y, nucleotide, ha='center', va='center', fontsize=8, color='black')

# Crear el gráfico
plt.figure(figsize=(20, len(lines) * 1.5))  # Ajustar tamaño según número de líneas
for line_num, line in enumerate(lines):
    for pos, nucleotide in enumerate(line):
        x = pos
        y = num_lines - line_num  # La línea más alta tiene Y=1
        color = nucleotide_colors.get(nucleotide, 'gray')  # Por si hay caracteres no esperados
        # Dibujar cuadro
        plt.gca().add_patch(plt.Rectangle((x - 0.5, y - 0.5), 1, 1, color=color, edgecolor='black'))
        # Etiquetar cuadro con el nucleótido
        plt.text(x, y, nucleotide, ha='center', va='center', fontsize=8, color='black')

# Configurar ejes
plt.xticks(range(0, line_length, 10), labels=range(1, line_length + 1, 10), fontsize=8)
plt.yticks(range(1, num_lines + 1), labels=[f"Línea {num_lines - i}" for i in range(num_lines)], fontsize=10)
plt.xlabel("Posición en la línea", fontsize=12)
plt.ylabel("Líneas de la secuencia", fontsize=12)
plt.title("Representación de secuencia de nucleótidos con etiquetas", fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Ajustar límites
plt.xlim(-0.5, line_length - 0.5)
plt.ylim(0.5, num_lines + 0.5)

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

```python
from collections import Counter
import matplotlib.pyplot as plt


# Contar frecuencia de nucleótidos
nucleotide_counts = Counter(sequence)

# Crear gráfico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(nucleotide_counts.keys(), nucleotide_counts.values(), color=['forestgreen', 'khaki', 'lightpink', 'lightskyblue'])

# Agregar etiquetas a cada barra
for bar in bars:
    height = bar.get_height()  # Obtener altura de la barra
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.5,  # Posición de la etiqueta
             f"{int(height)}", ha="center", va="bottom", fontsize=12, color="black")

# Personalizar gráfico
plt.xlabel("Nucleótidos")
plt.ylabel("Frecuencia")
plt.title("Composición de nucleótidos en la secuencia")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar el gráfico
plt.show()
```
