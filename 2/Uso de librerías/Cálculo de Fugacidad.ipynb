# Ejemplo
## Determinar la fugacidad del 1-buteno a 200¬∞C y 70 bar

```python
# Datos
T = 200+273.15 # Kelvin
P = 70 # bar
```

```python
# Tabla B1, Smith/Van Ness
w = 0.191
Pc = 40.43 # bar
Tc = 420 # K
```

```python
# variables reducidas
Pr = P/Pc
Tr=T/Tc
```

```python
# Definir par√°metros B¬∞ y B'
Bo = 0.083-0.422/Tr**1.6
B1=0.139-0.172/Tr**4.2

print(f"El valor calculado de Bo = {Bo}")
print(f"El valor calculado de B1 = {B1}")
```

# Calcular el coeficiente de fugacidad:
$$\phi=exp(\frac{Pr}{Tr}) (B^0+wB^1)$$

```python
import math
phi=math.exp((Pr/Tr)*(Bo+w*B1))

print(f"El valor calculado de coeficiente de fugacidad es {phi}")
```

# Comparaci√≥n con el valor calculado por Lee/Kesler
## Video explicaci√≥n

https://www.youtube.com/watch?v=hutI7HAvvRs

```python
#Comparaci√≥n con tablas de Lee/Kessler
phi_LK=0.638601

er = abs(phi-phi_LK)/(phi_LK)*100

print(f"el % de error con respecto al valor calculado \ndel coeficiente de fugacidad con correlaci√≥n de Lee Kessler es")
print(f"{er} %")
```
