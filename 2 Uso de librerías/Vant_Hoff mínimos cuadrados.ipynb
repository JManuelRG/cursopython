```python
# Ingresar los datos:
T=[800,900,1000,1100] 
K=[0.00103,0.0234,0.29,2.5] 
```

image.png

```python
# Evaluar los valores de la ec. de van't hoff
x=[] # 1/T
y=[] # ln K
for i in range(len(T)):
  x.append(1/T[i])


import math
for i in range(len(K)):
  y.append(math.log(K[i]))
```

```python
x
```

```python
y
```

```python
# Determinar valores promedio
xprom=(1/len(x))*sum(x)
yprom=((1/len(y))*sum(y))

print(f"El valor promedio de X es {xprom}")
print(f"El valor promedio de Y es {yprom}")
```

```python
num=[]
denom=[]
for i in range(len(x)):
  num.append(x[i]*(y[i]-yprom))
  
for i in range(len(y)):
  denom.append(x[i]*(x[i]-xprom))
```

```python
num
```

```python
denom
```

```python
# Cálculo de la pendiente
m = sum(num)/sum(denom)

print(f"El valor de la pendiente es: {m}")
```

```python
# Cálculo del intercepto
b=yprom-m*xprom

print(f"El valor del intercepto es: {b}")
```

```python
print(f"a) La ec. ajustada es: lnK = {m}*(1/T)+{b}")
```

```python
# Entalpía estándar de reacción
R=8.314 #J/molK
dH=-m*R

print(f"b) Entalpía estándar de reacción es: {-m*R} J/mol")
```

```python
# Cálculo de entropía
dS=b*R

print(f"b) Entropía estándar de reacción es: {b*R} J/molK")
```

```python
# K a 1200 K
print(f"c) El valor de K a T=1200 K es: {math.exp(m*(1/1200)+b)}")
```

```python
# Método directo
# librería scipy
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html

# importar librería
from scipy.stats import linregress

pendiente, intercepto, r, pvalor, desvest = linregress(x, y)

print(f"Resultados: pendiente={pendiente}, intercepto={intercepto}")
```
