```python
import sympy as sp
T = sp.symbols('T')
```

```python
T
```

```python
A=3.47
B=1.45e-3
C=0
D=0.121e5
Cp = A+B*T+C*T**2+D*T**-2
display(Cp)
```

```python
Cp.subs(T,298.15)
```

```python
R = 8.314 #J/molK
print(f"Cp(vapor agua) = {Cp.subs(T,400)*R} J/molK")
```

```python
Tsat=90+273.15
T2=130+273.15
```

```python
sp.integrate(Cp,(T,Tsat,T2))
```

```python
Qsens = (50)*(R)/18e-3*sp.integrate(Cp,(T,Tsat,T2))/1000
print(f"Qsens = {Qsens} kJ")
```

```python
Qlat = 114125 #kJ
```

```python
Qtotal = Qsens + Qlat
print(f"Qtotal = Qsens + Qlat = {Qsens} kJ + {Qlat} kJ = {Qtotal} kJ")
```
