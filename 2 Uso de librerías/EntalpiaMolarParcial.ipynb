```python
from sympy import *
```

```python
# Definir variables independientes
x1 = symbols('x_1')
x1
```

```python
x2 = 1-x1
x2
```

```python
H = 9*x1*x2+x1
H
```

```python
H.simplify()
```

```python
# Entalp√≠a molar parcial 1
```

# LATEX
#$\bar{H}_1=(\frac{\partial{H}}{\partial{n_1}})_{P, T, n_2}$

```python
Hmp1 = H+x2*H.diff(x1)
Hmp1.simplify()
```

```python
Hmp1.subs(x1,0.6)
```

```python
Hmp2 = H-x1*H.diff(x1)
Hmp2.simplify()
```

```python
Hmp2.subs(x1, 0.6)
```

# Entalp√≠a Molar Mezcla
# $$H=9x_1x_2+x_1$$

# Entalp√≠a Molar Parcial 1
#$$\bar{H_1}=(10x_1-9x_1^2)+x_2(10-18x_1)$$

# Entalp√≠a Molar Parcial 2
#$$\bar{H_2}=(10x_1-9x_1^2)-x_1(10-18x_1)$$

```python
import numpy as np
```

```python
x1=np.linspace(start=0,stop=1,num=100)
x2=1-x1
Hmezcla=10*x1-9*x1**2
H1mp=10*x1-9*x1**2+(1-x1)*(10-18*x1)
H2mp=10*x1-9*x1**2-x1*(10-18*x1)
```

```python
x1
```

```python
import matplotlib.pyplot as plt
plt.plot(x1,Hmezcla, label='$H$')
plt.plot(x1,H1mp, label='$\overline{H}_1$')
plt.plot(x1,H2mp, label='$\overline{H}_2$')
plt.plot(x1[0],H1mp[0], 'go', label='$\overline{H}_1^{\infty}$')
plt.plot(x1[99],H2mp[99], 'ro', label='$\overline{H}_2^{\infty}$')
plt.legend()
plt.title('Propiedades de la mezcla')
plt.xlabel('Composici√≥n $x_1$')
plt.ylabel('Entalp√≠a (kJ/mol)')
plt.grid()
plt.show()
```

```python
# Tabla de Datos
import pandas as pd
# Conversi√≥n a DataFrame
df = pd.DataFrame({"x1": x1,
                   "x2": x2,
                   "H (kJ/mol)":Hmezcla,
                   "H1mp (kJ/mol)":H1mp,
                   "H2mp (kJ/mol)":H2mp})

# Resultado
df.head()
```
