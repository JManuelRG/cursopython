```python
from sympy import *
T=symbols('T')

```

```python
print("Cp agua l√≠quida:")
Cpal=75.4e-3 #kJ/molK
display(Cpal)

print("Cp agua vapor")
Cpav=33.46e-3+0.688e-5*T+0.7604e-8*T**2-3.593e-12*T**3
display(Cpav)

print("Cp agua hielo")
cph=2.090*0.018
display(cph) #kJ/(mol K)
```

```python
h1 =-7.96e-3
```

```python
h2 = cph*(0--50)
h2
```
