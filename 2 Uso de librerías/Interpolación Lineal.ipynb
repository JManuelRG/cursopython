# Interpolación Lineal en Python.

La interpolación es un método para encontrar el valor $ y $ de una función $y=f(x)$, asumiendo que ese valor se encuentra entre dos puntos conocidos con coordenadas $ P(x_o,y_o) $ y $Q(x_1,y_1)$.

El método asume que existe una tendencia lineal entre los puntos $ P(x_o,y_o) $ y $Q(x_1,y_1)$, por lo que las pendientes entre los puntos son iguales.

# Implementación en Python:

Puede copiar y pegar las siguientes líneas de código, cambiando los valores de los puntos $ P(x_o,y_o) $, $Q(x_1,y_1)$ y $x$, este programa realizará la interpolación y mostrará el valor interpolado $y$.

```python
x0=1.1
x1=1.15
y0=0.99
y1=1.4
x=1.127

y=(y1-y0)/(x1-x0)*(x-x0)+y0
print("El valor interpolado es:")
print(y)
```
