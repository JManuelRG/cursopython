```python
H7 = Cpal*(100-25)+40.656+integrate(Cpav,(T,100,1000))
print(f"H7 = {Cpal*(100-25)} kJ/mol + 40.656  kJ/mol + {integrate(Cpav,(T,100,1000))} kJ/mol")
print(f"H7 = {H7} kJ/mol")
```

#$$\dot {Q}=\dot{\Delta} H = \xi \Delta H_R°+\sum \dot{n}_{salidas}\hat{H}_{salidas}-\sum \dot{n}_{entradas}\hat{H}_{entradas}$$

```python
# Datos
dHr=-2220.06 #kJ/mol calor de reacción
E=100 #mol grado de avance

# Entalpías en kJ/mol:
H1 = 0
H2 = 8.47
H3 = 8.12
H4 = 32.47
H5 = 30.56
H6 = 48.6

#Sustitución
Q = (dHr)*E  + 100*H4 + 2256*H5 + 300*H6 + 400*H7  - 100*H1 - 600*H2  - 2256*H3 #Entradas

# Respuesta:
print("La cantidad de calor transferido en el horno es de:")
print(f"{Q} kW")
```
