```python
from sympy import * #Librería para resolver integrales
T=symbols('T')
```

```python
print("Cp agua líquida:")
Cpal=75.4e-3 #kJ/molK
display(Cpal)

print("Cp agua vapor")
Cpav=33.46e-3+0.688e-5*T+0.7604e-8*T**2-3.593e-12*T**3
display(Cpav)
```

```python
# Entalpía de evaporación del agua
Hev = 40.656
H7 = Cpal*(100-25)+40.656+integrate(Cpav,(T,100,1000))
print(f"H7 = {Cpal*(100-25)} kJ/mol + {Hev}  kJ/mol + {integrate(Cpav,(T,100,1000))} kJ/mol")
print(f"H7 = {H7} kJ/mol")
```

# Resolución del balance de energía

#$$\dot {Q}=\dot{\Delta} H = \xi \Delta H_R°+\sum \dot{n}_{salidas}\hat{H}_{salidas}-\sum \dot{n}_{entradas}\hat{H}_{entradas}$$

#Tabla de Entalpías
image.png

```python
# Datos
dHr=-2220.06 #kJ/mol calor de reacción
E=100 #mol grado de avance


# Entalpías en kJ/mol:
# Datos de Tabla B8 del libro de Felder, también se pueden calcular
# con la integral de Cp de cada gas desde 25°C hasta la Tfinal de cada flujo
H1 = 0
H2 = 8.47
H3 = 8.12
H4 = 32.47
H5 = 30.56
H6 = 48.6

#Sustitución
Q = (dHr)*E  + 100*H4 + 2256*H5 + 300*H6 + 400*H7  - 100*H1 - 600*H2  - 2256*H3 #Entradas

# Respuesta:
print("La cantidad de calor transferido en el horno es de:")
print(f"{Q} kW")
```
