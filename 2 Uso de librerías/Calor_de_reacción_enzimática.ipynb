$$\Delta H°_{R309.15=}\Delta H°_{R298.15}+R\int_{298.15}^{309.15}\frac{C_p}{R}dT$$


```python
dA=-7.625
dB=0.001089
dC=-4.42e-6
dD=103600
```

```python
import sympy
T = sympy.symbols('T')
CpR=dA+dB*T+dC*T**2+dD*T**-2
display(CpR)
```

```python
#Datos
dH298= 21327 # J/mol
R=8.314 #J/molK

```

$$\Delta H°_{R309.15=}\Delta H°_{R298.15}+R\int_{298.15}^{309.15}\frac{C_p}{R}dT$$

$$\Delta H°_{R309.15}=21327 J/mol+R\int_{298.15}^{309.15} {{-4.42x10^{-6}T^2+0.001089T-7.624+\frac{103600}{T^2}}} dT$$

```python
integral = sympy.integrate(CpR,T)
display(integral)
```

```python
integral = sympy.integrate(CpR,(T,298.15,309.15))
display(integral)
```

```python
# Calor de reacción
calor = dH298+R*integral
print(f"El calor de reaccion a 309.15 K es de {calor} J/mol")
```
