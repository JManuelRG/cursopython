```python
dA=1.08E+04	
dB=2.27E+00
dD=-2.68E+07
```

```python
import sympy
```

```python
T=sympy.symbols('T')
display(T)
```

```python
Cp=dA+dB*T+dD*T**-2
display(Cp)
```

```python
integral=sympy.integrate(Cp,(T,298.15,723.15)) # molK/s
```

```python
integral #molK/s
```

```python
sympy.integrate(Cp,T)
```

```python
dHr298=-213119395
```

```python
R=8.314 #J/molK
R=1.987 BTU/lbmolR
Q = dHr298+R*integral
```
