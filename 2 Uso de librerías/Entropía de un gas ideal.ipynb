```python
A = 1.702
B = 9.081e-3
C =-2.164e-6
To =550 #K
Po =5 #bar
P = 1 #bar
D=0
```

```python
import sympy
T = sympy.symbols('T')
display(T)
type(T)
```

```python
CpR = A+B*T+C*T**2+D*T**-2
display(CpR)
```

```python
# Integral de Cp/TdT
integral_cp = sympy.integrate(CpR/T, (T, To, T)) 
display(integral_cp)
```

```python
import math
logaritmo=math.log(P/Po)
ecuacion = sympy.Eq(integral_cp - logaritmo,0)
display(ecuacion)
```

```python
# Definir funcin para graficar
# f(T)=0
f=-1.082e-6*T**2+0.009081*T-1.702*sympy.log(T)-10.73948-3.0578070875659
display(f)
```
