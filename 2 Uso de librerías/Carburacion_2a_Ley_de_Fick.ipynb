```python
import numpy as np
T = np.linspace(start=1173.15,
                stop=1600,
                num=1000)
```

```python
import math

t = (0.01/(0.23*0.4969))*np.exp(16557.6/T)
```

```python
import matplotlib.pyplot as plt
plt.plot(T-273.15, t/3600)
plt.grid()
plt.xlabel('T(°C)')
plt.ylabel('t(horas)')
plt.title('Gráfico de Diseño (Difusión)')
```
