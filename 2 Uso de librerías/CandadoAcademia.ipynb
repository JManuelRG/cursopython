# Caso 1, uso de candado actual

```python
import numpy as np
cal_ex=np.linspace(0,100,101)
```

```python
def calcandado(ex, tareas):
    if ex<50 and ex>=40:
        tareas=tareas*0.9
    elif ex<40 and ex>=30:
        tareas=tareas*0.8
    elif ex<30 and ex>=20:
        tareas=tareas*0.7
    elif ex<20 and ex>=10:
        tareas=tareas*0.6
    elif ex<10:
        tareas=tareas*0
    final=ex*0.7+tareas

    return {'examen':ex,
            'tareas':tareas,
            'total100':final,
            'total10':round(final/10,2)}
```

```python
calcandado(50,30)
```

```python
vcand10=[]
vcand20=[]
vcand30=[]
for i in cal_ex:
    vcand10.append(calcandado(i, 10))
    vcand20.append(calcandado(i, 20))
    vcand30.append(calcandado(i, 30))
```

```python
import pandas as pd
Tablacand30= pd.DataFrame(vcand30)
Tablacand20= pd.DataFrame(vcand20)
Tablacand10= pd.DataFrame(vcand10)
Tablacand30
```

```python

import matplotlib.pyplot as plt
fig1 = plt.figure()

ax = fig1.add_subplot(1, 1, 1)
ax.plot(Tablacand30['examen'],
         Tablacand30['total100'], label='Con 30pts de actividades', color='blue')
ax.plot(Tablacand20['examen'],
         Tablacand20['total100'], label='Con 20pts de actividades', color='green')
ax.plot(Tablacand10['examen'],
         Tablacand10['total100'], label='Con 10pts de actividades', color='orange')
ax.axhline(y = 60, color = 'r', linestyle = '-',
          label='Aprobatoria')
ax.set_title('Calificaciones con Candado Actual')
ax.set_xlabel('Calificación del Examen')
ax.set_ylabel('Calificación Final')
# Major ticks every 20, minor ticks every 5
major_ticks = np.arange(0, 101, 5)
minor_ticks = np.arange(0, 101, 5)

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

# And a corresponding grid
ax.grid(which='both')

# Or if you want different settings for the grids:
ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),
          fancybox=True, shadow=True, ncol=4)
plt.show()
```

# Simulación del índice de reprobación a partir de la media y desviación estándar de un grupo
# Generando distribución normal y aplicando la ponderación

```python
mu, sigma = 45, 5 # mean and standard deviation
s = np.random.normal(mu, sigma, 35)
# generar 35 calificaciones
```

```python
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```
