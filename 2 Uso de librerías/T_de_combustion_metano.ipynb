# Balance de energía

```python
# Balance de energía
# Entalpías de reacción en kJ/mol
dHr1=-519.33
dHr2=-802.31

#Flujos en mol/h, se multiplicó por 1000 cada uno para convertir kmol/h a mol/h
#Grados de avance
e1=10000
e2=35000

#Entradas:
n1=50000 #CH4
n2=85000 #O2
n3=319760 #N2

# salidas
n4= 5000 #CH4
n7=319760 #N2
n6=35000 #CO2
n5=10000 #CO
n8=90000 #H2O
```

$$\dot{Q}=-8061 kW(\frac{3600s}{1h})$$

```python
Q=-8061*3600 
print(f"Calor liberado = {Q} kJ/h")
```

```python
# Ecuacion
ecuacion = Eq(Q,e1*dHr1 + e2*dHr2 + n4*H4 + n7*H7 + n6*H6 + n5*H5 + n8*H8 - n1*H1 - n2*H2 - n3*H3)
```

```python
print("Ecuación a Resolver:")
display(ecuacion)
```

```python
temperatura=solve(ecuacion,T)
temperatura
```

```python
print(f"La temperatura del reactor es de {temperatura[1]}°C")
print("Fin :D")
```
