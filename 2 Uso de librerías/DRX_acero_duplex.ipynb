# C√°lculo de los planos de difracci√≥n de austenita

# Ley de Bragg

#$d=\frac{1.54 A}{2sin\theta}$

```python
# funci√≥n de distancia interplanar
import math
def dist(angulo):
  angulo = angulo*math.pi/180
  d = 1.54/(2*math.sin(angulo/2))
  print("El valor de la distancia interplanar para un √°ngulo")
  print(f"2theta = {angulo} rad, es de")
  print(f"d = {d} A")
```
