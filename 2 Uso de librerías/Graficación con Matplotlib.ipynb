Visualización de datos
Uso de la librería Matplotlib para graficación de datos,

tutoriales de matplotlib:
https://matplotlib.org/tutorials/index.html

Ejemplo:
    
    Graficación de funciones:
    
    Graficar las funciones: 
    y1 = sin(x)
    y2 = cos(x)

```python
# Importar librerías:
import matplotlib.pyplot as plt
import numpy as np

# Generar datos de X a graficar:
x = np.linspace(0, 10, 100) # (inicio, final, # de datos)
# print(x)
# print(len(x))

# Generar datos de Y a graficar:
y1 = np.sin(x)
# print(y1)
y2 = np.cos(x)

# Graficar
plt.plot(x, y1, label="Y1 = sin(x)") # y1 = sin(x)
plt.plot(x, y2, label="Y2 = cos(x)") # y2 = cos(x)
plt.legend()
plt.title("Gráfico de funciones")
plt.xlabel("Valores de X")
plt.ylabel("Valores de Y")
plt.grid()
plt.show()
```
