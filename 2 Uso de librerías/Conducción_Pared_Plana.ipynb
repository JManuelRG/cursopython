# Variaci√≥n de temperatura en la pared
 $$ T(x)=({\frac{T_2-T_1}{L}})x+T_1 $$

```python
# Variables

T1 = 120 #¬∞C
T2 = 50 #¬∞C
k=1.2 #W/mK
A=15 #m2
L=0.2 #m
```

```python
# Declarar funci√≥n definida por el usuario
def Temp(x):
  return ((T2-T1)/L)*x+T1
```

```python
Temp(0)
```

```python
Temp(L)
```

```python
#a)
Temp(L/2) #En¬∞C
```

# b) Transferencia de calor

$$\dot{Q} = kA (T_1-T_2)/L$$ 

```python
Q=k*A*(T1-T2)/L
print(Q) #En Watts
```

```python
# Perfil de temperatura
import matplotlib.pyplot as plt # Librer√≠a para graficar
import numpy as np # Librer√≠a para generar vectores

X = np.linspace(0,L,50)
```

```python
X
```

```python
T = Temp(X)
T
```

```python
#Perfil de temperatura en la pared

plt.plot(X,T, label="T vs x")
plt.legend()
plt.title("Perfil de temperatura")
plt.xlabel("Espesor, m")
plt.ylabel("Temperatura, ¬∞C")
plt.grid()
plt.show()
```
