```python
K1=29.3659
T1=298.15
```

```python
import numpy as np
T = np.linspace(start=0,
                stop=1000,
                num=1001)
```

```python
T
```

```python
T = T+273.15
T
```

```python
dH= -45792
dG=-8378
R = 8.314
K = K1 * np.exp(dH/R*(1/T1-1/T))
```

```python
K
```

```python
import matplotlib.pyplot as plt
plt.plot(T, K)
plt.title('Perfil de K: $C_2H_4+H_2O->CH_3CH_2OH$')
plt.xlabel("T en K")
plt.ylabel("K de equilibrio")
plt.grid()
```

```python
import matplotlib.pyplot as plt
plt.plot(1/T, np.log(K))
plt.title('Perfil de K: $C_2H_4+H_2O->CH_3CH_2OH$')
plt.xlabel("1/T en 1/K")
plt.ylabel("ln K de equilibrio")
plt.grid()
```
