# Gr√°fico de la funci√≥n

#$$V(t)=1.2 + 0.05t-0.00125t^2$$

```python
# prompt: genera un gr√°fico general de una funci√≥n usando matplotlib

import numpy as np
t = np.linspace(start=0, stop=60, num=299)
V = 1.2+0.05*t-0.0025/2*t**2

import matplotlib.pyplot as plt
plt.plot(t, V)
plt.xlabel('tiempo (s)')
plt.ylabel('V ($m^3$)')
plt.title('Comportamiento del volumen')
plt.grid()
plt.show()
```
