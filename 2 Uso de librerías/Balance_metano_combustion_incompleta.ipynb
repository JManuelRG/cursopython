```python
from sympy import *
T=symbols('T')
```

```python
print("Cp agua líquida:")
Cpal=75.4e-3 #kJ/molK
display(Cpal)

print("Cp agua vapor")
Cpav=33.46e-3+0.688e-5*T+0.7604e-8*T**2-3.593e-12*T**3
display(Cpav)
```

```python
H8 = Cpal*(100-25)+40.656+integrate(Cpav,(T,100,400))
print(f"H8 = {Cpal*(100-25)} kJ/mol + 40.656  kJ/mol + {integrate(Cpav,(T,100,1000))} kJ/mol")
print(f"H8 = {H8} kJ/mol")
```

```python
# Capacidad calorífica metano gas
Cpmet=34.31e-3+5.469e-5*T+0.3661e-8*T**2- 11e-12*T**3
display(Cpmet)
```

```python
H1=integrate(Cpmet,(T,25,400))
print(H1)
```

```python
# Balance de energía
# Entalpías de reacción en kJ/mol
dHr1=-802.31
dHr2=-519.33

#Flujos en mol/h, se multiplicó por 1000 cada uno para convertir kmol/h a mol/h
#Grados de avance
e1=10000
e2=35000

#Entradas:
n1=50000 #CH4
n2=85000 #O2
n3=319760 #N2

# salidas
n4= 5000 #CH4
n7=319760 #N2
n6=35000 #CO2
n5=10000 #CO
n8=90000 #H2O

#Entalpias en kJ/mol
H4=H1
H2=11.72
H3=H7=11.15
H6=16.35
H5=11.25
```

```python
# Cálculo del balance, calor en kJ/h
Q = e1*dHr1 + e2*dHr2 + n4*H4 + n7*H7 + n6*H6 + n5*H5 + n8*H8 - n1*H1 - n2*H2 - n3*H3
print("Calor liberado en el reactor:")
print(f"Q = {e1}*{dHr1} + {e2}*{dHr2} + {n4}*{H4} + {n7}*{H7} + {n6}*{H6} + {n5}*{H5} + {n8}*{H8} - {n1}*{H1} - {n2}*{H2} - {n3}*{H3} kJ/h")
```

```python
print(f"Q={Q} kJ/h")
print(f"Q={Q/3600} kw")
```
