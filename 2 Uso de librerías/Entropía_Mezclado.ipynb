```python
import sympy as sp # Librera de Clculo simblico
```

```python
T = sp.symbols('T')
```

```python
CpRN = 3.28 +0.593e-3*T+0.04e5*T**-2

display(CpRN)
```

```python
CpRO = 3.639 +0.506e-3*T-0.227e5*T**-2

display(CpRO)
```

```python
CpRair = 0.21*CpRO + 0.79*CpRN

display(CpRair)
```

image.png

```python
# Calor:
na= 1 #mol/s
nb = 2
R = 8.314 # J/molK

Q = na*R*sp.integrate(CpRair,(T, 600, 400)) + nb*R*sp.integrate(CpRair,(T, 450, 400))
print(f"Q = {Q} J/s")
```

```python
# Cambios de entropa
S = na*R*sp.integrate(CpRair/T,(T, 600, 400)) + nb*R*sp.integrate(CpRair/T,(T, 450, 400))
S = S.evalf()
print(f"S = {S} J/K")
```

```python
# Generacin de entropa:
Ta=300 #K
Sg = S - Q/Ta
print(f"Sg = {Sg} J/K")
```
