# Propiedades de la mezcla:

#$$V = V_f + X (V_g-V_f)$$
#$$H = h_f + X (h_g-h_f) = h_f + Xh_{fg} $$
#$$U = u_f + X (u_g-u_f) = u_f + Xu_{fg} $$
#$$S = s_f + X (s_g-s_f) = s_f + Xs_{fg} $$

image.png

```python
# Propiedades del sistema a 130°C
vf= 0.00107 #m3/kg
vg= 0.66808 #m3/kg
vfg = vg-vf # = 0.66701 m3/kg

uf= 546.1 #KJ/kg
ufg= 1993.4 #KJ/kg

hf = 546.38 #KJ/kg
hfg = 2173.7 #KJ/kg

Sf = 1.6346 #KJ/kgK
Sfg = 5.3919 # KJ/kg

X=0.4
```

```python
def M(Mf,Mfg,X):
  return Mf+X*Mfg
```

```python
M(vf,vfg,X) #m3/kg
```

```python
M(hf,hfg,X) #kJ/kg
```

```python
M(uf,ufg,X) #kJ/kg
```

```python
M(Sf,Sfg,X) # kJ/kgK
```
