# Ecuación de Antoine

Ec. de Antoine:
$$ln P(kPa)=A-\frac{B}{T(°C)+C}$$

Video Grabado y Explicación: https://youtu.be/2ErK0BItNZY

image.png

```python
A = 16.3872
B = 3885.7
C = 230.17
```

```python
import math

def Psat(T):
  P = math.exp(A-B/(T+C))
  return P
```

```python
Psat(0.01)
```

```python
Psat(5)
```

```python
import numpy as np
Tsat = np.linspace(start=5, stop=370, num=74)
Tsat
```

```python
Pcalc=[]

for i in range(len(Tsat)):
  print(f"Tsat={Tsat[i]}, Psat={Psat(Tsat[i])}")
  Pcalc.append(Psat(Tsat[i]))
```

```python
Pcalc
```

```python
import matplotlib.pyplot as plt
plt.plot(Tsat, Pcalc)
plt.title('Ec. de Antoine')
plt.xlabel('Temperatura (°C)')
plt.ylabel('$P^{sat}(kPa)$')
plt.grid()
```
