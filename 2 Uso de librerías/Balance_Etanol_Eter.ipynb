# Agua

```python
from sympy import *
T = symbols('T')
T
```

```python
Cpal = 75.4e-3 #kJ/molK
Cpav = 33.46e-3+ 0.688e-5*T+0.7604e-8*T**2-3.593e-12*T**3
print("Cp agua líquida en kJ/molK:")
display(Cpal)
print("Cp agua vapor en kJ/molK:")
display(Cpav)
```

```python
Heva = 40.656 # KJ/mol
```

```python
print(f"{integrate(Cpal, (T, 25,100)) } kJ/mol + {Heva} kJ/mol + {integrate(Cpav, (T, 100,350))}")
```

```python
integrate(Cpal, (T, 25,100)) + Heva + integrate(Cpav, (T, 100,350))
```
