```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import special

X = np.linspace(start=0, stop=2, num=100)
ERROR = special.erf(X)

plt.plot(X, ERROR)
plt.title('Función Error')
plt.xlabel('x')
plt.ylabel('Valor de Función Error')
plt.grid()

# Acotando el eje x en incrementos de 0.1
plt.xticks(np.arange(0, 2.1, 0.1), rotation=90)

# Acotando el eje y en incrementos de 0.1
plt.yticks(np.arange(0, 1.1, 0.1))

plt.show()
```
