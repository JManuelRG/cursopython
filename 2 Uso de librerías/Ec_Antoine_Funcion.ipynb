# $$P^{sat}= exp(A-{\frac{B}{T+C}})$$

```python
T = 30 #C
```

```python
import math

def Psat(A,B,C,T):
  Psat = math.exp(A-(B/(T+C)))
  return f"Psat = {Psat} kPa"
```
