Planteamiento del problema:

En una reacción química de cracking de una cantidad de n-butano puro a 750 K y 1.2 bar, se producen olefinas de acuerdo a las siguientes reacciones:
image.png



Los productos principales con etileno, etano, propileno y metano.

Si estas reacciones logran el equilibrio, ¿Cuál es la composición del producto?

Las constantes de equilibrio a esta temperatura se han calculado de acuerdo con el procedimiento descrito previamente.

KI = 3.856
KII = 268.4

Explicación:

https://www.youtube.com/watch?v=LF9D0qUl5oA

Sistema de ecuaciones a resolver:
image.png

```python
# Datos
P=1.2 #bar
Po=1 #bar
KI=3.856
KII=268.4
```

```python
from sympy import  *
e1 = symbols('epsilon_1') 
e2 = symbols('epsilon_2')
display(e1)
display(e2)
```

```python
ec1 = Eq(e1**2/((1-e1-e2)*(1+e1+e2)),(P/Po)**-1*KI)
display(ec1)
```

```python
ec2 = Eq(e2**2/((1-e1-e2)*(1+e1+e2)),(P/Po)**-1*KII)
display(ec2)
```

```python
solve([ec1,ec2],e1,e2)
```
