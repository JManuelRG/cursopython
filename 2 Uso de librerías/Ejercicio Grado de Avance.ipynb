# Resolver la ecuación, para obtener el grado de avance:

# Ecuación escrita en látex: https://www.latex-project.org/
$$59.4 = \frac{(3+2\epsilon)^2}{(8-\epsilon)(5-\epsilon)}$$


```python
# Usaremos la librería sympy: sympy.org
import sympy
```

```python
# Solución de ecuación no lineal

# 1. crear la variable independiente como variable simbólica
epsilon = sympy.symbols('epsilon')
display(epsilon)
```

```python
# 2. Definir la ecuación
ecuacion = sympy.Eq(59.4, (3+2*epsilon)**2/((8-epsilon)*(5-epsilon)))
display(ecuacion)
```

```python
# 3. Resolver la ecuación
solucion = sympy.solve(ecuacion, epsilon)
display(solucion)
```

```python
print("El grado de avance de la reacción es:")
solucion[0]
```

# Solución obtenida:
## Grado de avance:
$$\epsilon = 4.36338$$
