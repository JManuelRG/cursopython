# Resolver la ecuacin, para obtener el grado de avance:

# Ecuacin escrita en ltex: https://www.latex-project.org/
$$59.4 = \frac{(3+2\epsilon)^2}{(8-\epsilon)(5-\epsilon)}$$


```python
# Usaremos la librera sympy: sympy.org
import sympy
```

```python
# Solucin de ecuacin no lineal

# 1. crear la variable independiente como variable simblica
epsilon = sympy.symbols('epsilon')
display(epsilon)
```

```python
# 2. Definir la ecuacin
ecuacion = sympy.Eq(59.4, (3+2*epsilon)**2/((8-epsilon)*(5-epsilon)))
display(ecuacion)
```

```python
# 3. Resolver la ecuacin
solucion = sympy.solve(ecuacion, epsilon)
display(solucion)
```

```python
print("El grado de avance de la reaccin es:")
solucion[0]
```

# Solucin obtenida:
## Grado de avance:
$$\epsilon = 4.36338$$
