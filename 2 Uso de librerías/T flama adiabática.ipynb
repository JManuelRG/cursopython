```python
dA = 43.471
dB = 0.00950219
dC = 0
dD = -64460
dHr = -802625 # J/mol
R = 8.314 #J/mol
```

```python
import sympy
T = sympy.symbols('T')
```

```python
integral = sympy.integrate(dA+dB*T+dC*T**2+dD*T**-2, (T,298.15,T))
display(integral)
```

```python
ecuacion = sympy.Eq(dHr, -R*integral)
display(ecuacion)
```

```python
# Resolver para T
Tflama= sympy.solve(ecuacion,T)
print(Tflama)
```
