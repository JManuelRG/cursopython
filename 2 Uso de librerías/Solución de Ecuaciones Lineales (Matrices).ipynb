```python
import numpy as np
```

```python
A = np.array([[24,6,12],
              [-3,0,1],
              [-4,1,-4]])
print(A)
```

```python
b = np.array([156,
              0,
              0])
print(b)
```

```python
# Soluci√≥n
X = b*np.linalg.inv(A)
print(X)
```

```python
print("La soluci√≥n al sistema es:")
print(f"x = {X[0][0]}") 
print(f"y = {X[1][0]}")
print(f"z = {X[2][0]}")
# x = 1
```

```python
# soluci√≥n
from math import ceil
x=ceil(X[0][0])
y=X[1][0]
z=X[2][0]

print(f"1 L de leche cuesta {x} ‚Ç¨")
print(f"1 kg de jam√≥n cuesta {y} ‚Ç¨")
print(f"1 L de aceite de oliva cuesta {z} ‚Ç¨")
```

```python
# Comprobaci√≥n
# Ecuaci√≥n 1
print("Ecuaci√≥n 1: ", 24*x+6*y+12*z) # Debe dar 156

# Ecuaci√≥n 2
print("Ecuaci√≥n 2: z = ", 3*x) # Debe dar 3*x = 3

# Ecuaci√≥n 3
print("Ecuaci√≥n 3: y = ", 4*z+4*x) # Debe dar 16
```
