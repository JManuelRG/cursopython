# Transferencia de Calor por Convección
# Perfiles de velocidad y de temperatura en una chumacera


## Perfil de velocidad

$$u(y)=\frac {y}{L}V$$



# Planteamiento del problema:

https://www.youtube.com/watch?v=R7FYSlac79E



# Solución en este cuaderno
https://youtu.be/8J1uIMIyZD8

```python
# Construcción del perfil
import numpy as np # librería numérica
import matplotlib.pyplot as plt # librería de visualización de datos

# Espesor de la capa de aceite
L = 2/1000 #m

# Velocidad de la placa móvil
V = 12 #m/s

#vector y
y = np.linspace(0,L,100)

#velocidad
u=(y/L)*V
```

```python
# Graficar
plt.plot(u,y, label='Velocidad')
plt.title('Perfil de Velocidad')
plt.xlabel('Velocidad, m/s')
plt.ylabel('espesor, (m)')
plt.grid(axis='both')
plt.legend()
plt.show()
```

# Perfil de Temperatura
$$T(y)=T_0+ \frac {\mu V^2}{2k}(\frac {y}{L}-\frac {y^2}{L^2}) $$

```python
# Propiedades físicas
k=0.145 #W/mK
To=20 #°C
mu = 0.8374 #kg/ms

#Temperatura T(y):
T = To+(mu*V**2/(2*k)*(y/L-(y/L)**2))
```

```python
# Graficar
plt.plot(T,y, label='Temperatura')
plt.title('Perfil de Temperatura')
plt.xlabel('Temperatura, °C')
plt.ylabel('espesor, (m)')
plt.grid(axis='both')
plt.legend()
plt.show()
```

```python
#Transferencia de Calor
#Placa Superior
q0=-mu*V**2/(2*L) #W/m^2

#Placa Inferior
qL=mu*V**2/(2*L) #W/m^2
```

```python
print(f"El flujo de calor en la placa superior es {q0} W/m^2")
print(f"El flujo de calor en la placa inferior es {qL} W/m^2")
```
