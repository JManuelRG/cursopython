```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
T = np.linspace(start=900, stop=1400, num=501)
T=T+273.15
```

```python
R=1.987
t = ((0.1/0.703)**2)/(0.23*np.exp(-32900/(R*T)))/3600
```

```python
# Tabla de Datos
import pandas as pd
datos = {"T (K)":T,
       "t (h)":t}
Tabla =pd.DataFrame(datos)
print(Tabla)
```

```python
plt.plot(T,t)
plt.title('Condiciones del Tratamiento trmico')
plt.grid()
plt.xlabel('Temperatura K')
plt.ylabel('tiempo (h)')
```
