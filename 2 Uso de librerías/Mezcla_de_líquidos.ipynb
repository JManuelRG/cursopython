```python
T1=30 #°C
m1=200 #g
T2=80 #°C
m2=600 #g

Q=10000 #cal
C1=1 #cal/g°C
C2=1 #cal/g°C
```

```python
T3 = (Q+m1*C1*T1+m2*C2*T2)/(m1*C1+m2*C2)
print(f"El valor de la temperatura de equilibrio es {T3}°C")
```
