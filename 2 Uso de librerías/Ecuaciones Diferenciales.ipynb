# Cómo resolver ecuaciones diferenciales en Python
# Utilizando Sympy

```python
# IMPORTAR TODAS LAS FUNCIONES DE LA LIBRERÍA SYMPY
from sympy import *

# Visualización de resultados en forma de ecuación:
# Utilizará Látex para la escritura de fórmulas, si no se tiene látex, por defecto se usa MathJax
from sympy.interactive import printing
printing.init_printing(use_latex=True)
```

```python
# Imprimir línea vacía para espaciar fórmulas
def lin():
  print('\n') # imprime línea vacía
```

```python
# Cómo definir variables

# Definir variables independientes simbólicas:
x = symbols('x')
y = symbols('y')
z = symbols('z')

# alternativa: x,y=symbols('x y')
display(x)
display(y)
display(z)
```

```python
# Definir funciones (variables dependientes)
f = Function('f')('x')
g = Function('g')('y')
M = Function('M')('z')

display(f)
display(g)
display(M)
```

```python
# Definir Funciones de dos o más variables
h=Function('h')
h=h(x,y)
display(h)
```

```python
# Definir Primeras Derivadas

# Primera derivada, función diff de la librería sympy
df = diff(f,x)
dg = diff(g,y)
dhx = diff(h,x)
dhy = diff(h,y)

display(df)
lin() # Línea vacía
display(dg)
lin()
display(dhx)
lin()
display(dhy)
```

```python
# Definir segundas derivadas

# Segundas derivadas, función diff de la librería sympy
d2f = diff(f,x,x)
d2g = diff(g,y,y)
d2h = diff(h,y,x)

display(d2f)
lin()
display(d2g)
lin()
display(d2h)
lin()
```

# Ejemplo: Resolver la ecuación:
ec separables.jpg

```python
# Definir ecuación a resolver
x=symbols('x')
y=Function('y')('x')
ecuacion = diff(y,x)-1/x**2
display(ecuacion)
```

```python
# Obtener solución general
dsolve(ecuacion)
```

# Ejemplo de Solución 2
Ecuación de Flujo bidimensional en coordenadas cilíndricas en estado estacionario
Ec cilndro.jpg



```python
# Definir variables y Funciones
r=symbols('r')
T=Function('T')('r')
display(T)
lin()
#Definir Ecuación
ecuacion2 = diff(r*diff(T,r),r)
display(ecuacion2)
```

```python
# Resolver ecuación
dsolve(ecuacion2)
```

# Ejemplo: Ecuación de Laplace
image.png

```python
# Definir función
T = Function('T')
x,y=symbols('x y')
T=T(x,y)
display(T)
```

```python
# Definir ecuación
x,y=symbols('x y')
ecuacion3 = diff(T,x,x)+diff(T,y,y)
display(ecuacion3)
```

```python
pdsolve(ecuacion3)
# No se puede resolver de manera analítica por esta herramienta
```
