# Determinacin del azotropo

```python
# Clculo de presiones de saturacin
import math
T = 318.15
P1 = math.exp(16.59-3643.31/(T-33.424))
print(P1) #KPa
```

```python
P2 = math.exp(14.25326-2665.54/(T-53.424))
print(P2) #KPa
```

```python
#Definir variables simblicas
import sympy
x_1 = sympy.symbols('x1')
```

```python
x_1
```

```python
A = 2.771-0.00523*T
A
```

```python
# Funciones del coeficiente de actividad
gamma_1 = sympy.exp(A*(1-x_1)**2)

gamma_1
```

```python
# Funciones del coeficiente de actividad
gamma_2 = sympy.exp(A*(x_1)**2)

gamma_2
```

```python
ec = sympy.Eq(gamma_1/gamma_2, P2/P1)
ec
```

```python
ec2 = sympy.Eq(sympy.log(gamma_1/gamma_2), sympy.log(P2/P1))
ec2
```

```python
#Composicin del azotropo: x1
sympy.solve(ec2,x_1)
```
