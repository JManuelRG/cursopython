# Cálculo del Calor de Reacción
image.png

```python
# Paso 1
# Symbolic Python
import sympy 
# Definir variable independiente
T = sympy.symbols('T')
display(T)
```

```python
CpR = -7.663+0.010815*T-3.5e-6*T**2-13500*T**-2
display(CpR)
```

```python
# Cálculo de la integral de Cp/R
integral = sympy.integrate(CpR, (T,298.15,1073.15))
integral # K
```

```python
# Calor de la reacción a 1073.15 K
# en J/mol

-90135+8.314*integral 
```
