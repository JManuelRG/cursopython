$$Q/n =R \int_{T_1}^{T_2}{C_P/R} dT$$

# a) Usando la tabla del polinomio de Cp

```python
import sympy #Symbolic Python

T = sympy.symbols('T')

CpO = 3.639+.506e-3*T-.227e5*T**(-2)
CpN =3.28+.593e-3*T+0.04e5*T**(-2)
```

```python
display(CpO)
```

```python
display(CpN)
```

$$(C_P/R)_M = \sum_i{(C_P/R)_iy_i}$$

```python
sympy.Eq(0.21*CpO+0.79*CpN,0)
```

```python
CpM = 0.21*CpO + 0.79*CpN
display(CpM)
```

Integral:
$$$$

```python
solucion = 8.314*sympy.integrate(CpM, (T, 320, 260))
print(solucion) # En J/mol
```

a) Resultado = 
$$\dot{Q}/\dot{n} = -1747.31 J/mol \rightarrow kJ/kmol$$

# b) Usando Tablas de Propiedades de $$N_2 \quad y \quad O_2$$

$$\dot{Q}/\dot{n} = \Delta H = H_2-H_1$$

$$H_2 = Entalpía \quad a \quad T_2=260 K$$ 
$$H_1 = Entalpía \quad a \quad T_1=320 K$$ 

## Leyendo de tablas A18 y A19 Cengel

```python
# Entalpía del N2, Tabla A18


HN1 =  9306  # a 320 K en kJ/kmol
HN2 =  7558  # a 260 K en kJ/kmol
```

```python
# Entalpía del O2, Tabla A19

HO1 =  9325  # a 320 K en kJ/kmol
HO2 =  7566  # a 260 K en kJ/kmol
```

# Entalpía del aire

$$\dot{Q}/\dot{n}=\Delta H_M=\sum_i{y_i\Delta H_i}$$

```python
# Entalpía del aire

HM = 0.79*(HN2-HN1) + 0.21*(HO2-HO1)
HM
```

# b) Resultado:
$$\dot{Q}/\dot{n}=-1750.31 kJ/kmol$$

# c) Usando Tablas de Aire

$$\dot{Q}/\dot{n} = \Delta H = H_2-H_1$$

## Para el aire:
