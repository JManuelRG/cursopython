# Solución de la ecuación por el Método de Newton-Raphson

## Instrucciones
1. Ejecutar la Celda No. 2
2. Introducir valor supuesto, tolerancia y número de iteraciones.
3. Visualizar la tabla de resultados e iteraciones.

```python
# CELDA No. 2.
#Algoritmo del Método de Newton Raphson de solución de ecuaciones no lineales

#Entrada de datos:
#Aproximación inicial
Xsup = float(input(f"Introduce el valor supuesto de la variable: {var}sup = "))
tol=float(input("Introduce el valor de la tolerancia: = ")) #Tolerancia
N= float(input("No. de iteraciones máximas: = ")) #Iteraciones máximas
i=1 #Inicializar Iteraciones

#Función a resolver:
print(f"Función a resolver: f({var}) = ")
display(f)

#Derivada de la función f'(x)=D
print(f"Derivada de la función introducida: f'({var}) = ")
derivada = diff(f,x)
display(derivada)

#vectores de datos de la variable independiente para el gráfico
vectorx=[Xsup]
vectory =[f.subs(var,Xsup)]
#Tabla de datos
print("%3s %10s %10s %10s %10s" %("It",var+"sup","f("+var+"sup)","f'("+var+"sup)",var+"calc"))
while i<N:
  Xcalc=Xsup-f.subs(var,Xsup)/derivada.subs(var,Xsup)
  vectorx.append(Xcalc)
  vectory.append(f.subs(var,Xcalc))
  if abs((Xcalc-Xsup)/Xsup)<tol:
    print("%3d %10.5f %10.5f %10.5f %10.5f " %(i,Xsup,f.subs(var,Xsup),derivada.subs(var,Xsup),Xcalc))
    print(f"La solución a la ecuación es {var} = %8.5f en %d iteraciones" %(Xcalc, i))
    break
  else:
    print("%3d %10.5f %10.5f %10.5f %10.5f " %(i,Xsup,f.subs(var,Xsup),derivada.subs(var,Xsup),Xcalc))
    i+=1
    Xsup=Xcalc

if i==N:
  print(f"El método no encontró la solución en {i} iteraciones. Pruebe lo siguiente: \n")
  print("Aumentar las iteraciones \n")
  print("Disminuya la tolerancia \n")
  print("Cambie el valor supuesto \n")
  print("Verifique si la función tiene una solución real \n")
  print("Vuelva a ejecutar esta celda \n")
```

# Gráfica del método

## Instrucciones

1. Ejecutar la celda siguiente para obtener la visualización gráfica del método

```python
# importar librería
import matplotlib.pyplot as plt
import numpy as np

# Generar gráfico del método y de las iteraciones
# vector variable independiente

# Generar valores a graficar

# Límites en la variable independiente
limmin = min(vectorx)
limmax = max(vectorx)

# Generar gráfico
valoresx = np.linspace(start=float(limmin*0.9),
                       stop=float(limmax*1.1),
                       num=100)
# Valores de a función:
valoresy=[]
for i in range(len(valoresx)):
  valoresy.append(f.subs(var,valoresx[i]))

#Generación del Gráfico
#Curva principal:
plt.plot(valoresx,valoresy,label=f"Función f({var})")

#iteraciones del método de Newton Raphson
for i in range(len(vectorx)): 
  plt.plot([vectorx[i],vectorx[i]], 
          [min(valoresy),vectory[i]],
          label=f"Iteración {i}: {var}={round(vectorx[i],4)}, f({var})={round(vectory[i],4)}") # línea vertical
  if i!=len(vectorx)-1:
    plt.plot([vectorx[i],vectorx[i+1]],
            [vectory[i],min(valoresy)],
            label=f"Iteración {i}: f'({var})={round(derivada.subs(var,vectorx[i]),4)}") # Línea diagonal

# Rótulos del gráfico:
plt.title(f'Gráfico de la función {f}')
plt.xlabel(f'Valores de la variable {var}')
plt.ylabel(f'Valores de la función')
plt.legend(loc='upper left',shadow=True, bbox_to_anchor=(1,1))
plt.grid(True)
plt.show()
```
