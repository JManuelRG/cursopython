Devuelven un valor binario de una comparaci√≥n, son operadores compuestos (trabajan con dos o m√°s comparaciones y devuelven un resultado, de tipo binario)

```python
# Negaci√≥n

print(45 == 34) # Falso
not 45 == 34 # Verdadero
```

```python
# Operador Y
# Solo devuelve verdadero si ambas condiciones se cumplen
# En caso contrario, devuelve el valor Falso

print(4==5 and 5>8)
# Falso y Falso = Falso
```

```python
print(4==5 and 5<8)
# Falso y Verdadero = Falso
```

```python
print(5==5 and 5<8)
# V y V = V
```

```python
# Operador O L√≥gico
# Devuelve verdadero, mientras una de las comparaciones sea verdadera
(3==3) or (23!=23)
# V o F = V
```
