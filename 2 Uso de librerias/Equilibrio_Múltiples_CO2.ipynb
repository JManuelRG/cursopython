```python
from sympy import *
```

```python
e1=symbols('\epsilon_1')
e2=symbols('\epsilon_2')
```

```python
e2
```

image.png

```python
yCO2=(2-e1-e2)/(8-2*e1)
yCO2
```

```python
yH2 = (5-3*e1-e2)/(8-2*e1)
yH2
```

```python
ymet = e1/((8-2*e1))
ymet
```

```python
yCO=(1+e2)/(8-2*e1)
yCO
```

```python
yH2O=(e1+e2)/(8-2*e1)
yH2O
```

```python
ec1 = Eq( 6.8e-5 ,ymet*yH2O / ((yCO2)*(yH2)**3))
ec1
```

```python
ec2=Eq( 8.4e-3 ,yCO*yH2O / ((yCO2)*(yH2)))
ec2
```

```python
sol=nsolve([ec1,ec2],
           [e1,e2],
           [.5,.5])
sol
```

```python
# Grado de avance reacción 1
g1=sol[0]
g1
```

```python
# Grado de avance reacción 2
g2=sol[1]
g2
```

```python
# Comprobación ec. 1
K1=ymet*yH2O / ((yCO2)*(yH2)**3)
K1
```

```python
K1.subs(e1,g1).subs(e2,g2)
```

```python
# Comprobacion ec2
K2=yCO*yH2O / ((yCO2)*(yH2))
K2
```

```python
K2.subs(e1,g1).subs(e2,g2)
```

# La igualdad se cumple en ambas ecuaciones
## Solución
## $\xi_1=3.3x10^{-3} mol$
## $\xi_1= 0.071 mol$

# Conversión de dióxido de carbono (reactivo limitante)
## $conversion = \frac{moles.consumidos}{moles.iniciales}$
## $conversion = \frac{\xi_1+\xi_2}{2}$


```python
# Conversión
(g1+g2)/2
```

# Fracciones molares en el equilibrio

```python
# Función para sustituir grado de avance en ecuaciones de fracciones mol
def sust(ecuacion):
  return ecuacion.subs(e1,g1).subs(e2,g2)
```

```python
# Fracciones molares en el equilibrio
# CO2
sust(yCO2)
```

```python
#Hidrógeno
sust(yH2)
```

```python
# CO
sust(yCO)
```

```python
#Agua
sust(yH2O)
```

```python
#Metanol
sust(ymet)
```
