```python
from scipy.optimize import root
import math
import numpy

def ec(D):
    return (-2*math.log10(5.4e-7/D + 2.5697e-5/D**3.5) - 0.2757/D**2.5)
Resp = root(ec, .5)
print(Resp)
```

# Resultados

```python
#Datos
nu=1.655e-5

#Resultados
D = round(Resp.x[0],8)
V=4*0.35/(math.pi*D**2)
A=math.pi*4*D**2
Re=V*D/nu
f=13.1587*D**5
```

```python
print("RESULTADOS")
print(f"El valor calculado del diámetro es de {D} m")
print(f"V = {V} m/s")
print(f"A={A} m2")
print(f"Re = {Re}")
print(f"f={f}")
```

# Solución Gráfica

```python
import matplotlib.pyplot as plt

vectorf=numpy.linspace(0.2,0.5,num=100)
vectorD=[]
for i in range(len(vectorf)):
  vectorD.append(ec(vectorf[i]))

plt.plot(vectorf,vectorD)
plt.title('Ecuación de Colebrook: Solución Gráfica')
plt.xlabel('diametro, m')
plt.ylabel('f(D)')
plt.grid()
```
